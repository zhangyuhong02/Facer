```python
intrinsics = torch.from_numpy(np.array(
        [[1100.0, 0, 225],
         [0, 1100.0, 225],
         [0, 0, 1]], dtype=np.float32)).unsqueeze(0)
intrinsics = torch.from_numpy(np.array(
        [[233, 0, 256],
         [0, 233, 256],
         [0, 0, 1]], dtype=np.float32)).unsqueeze(0)

resolution = 64

N = 1 # batch size
M = 4096 # ray number 
```

```python
  "focal_len": 1100.0,
  "cx": 225.0,
  "cy": 225.0,
```

```python
fx = intrinsics[:, 0, 0]
fy = intrinsics[:, 1, 1]
cx = intrinsics[:, 0, 2]
cy = intrinsics[:, 1, 2]
sk = intrinsics[:, 0, 1]

# change to transforms_train.json

uv = torch.stack(torch.meshgrid(torch.arange(resolution, dtype=torch.float32,), 
                                torch.arange(resolution, dtype=torch.float32,), 
                                indexing='ij'))# * (1./resolution) + (0.5/resolution)

uv = uv.flip(0).reshape(2, -1).transpose(1, 0)
uv = uv.unsqueeze(0).repeat(N, 1, 1)

x_cam = uv[:, :, 0].view(N, -1)
y_cam = uv[:, :, 1].view(N, -1)
z_cam = torch.ones((N, M), )

x_lift = (x_cam - cx.unsqueeze(-1) + cy.unsqueeze(-1)*sk.unsqueeze(-1)/fy.unsqueeze(-1) - sk.unsqueeze(-1)*y_cam/fy.unsqueeze(-1)) / fx.unsqueeze(-1) * z_cam
y_lift = (y_cam - cy.unsqueeze(-1)) / fy.unsqueeze(-1) * z_cam

cam_rel_points = torch.stack((x_lift, y_lift, z_cam, torch.ones_like(z_cam)), dim=-1)
```

```python
cam_rel_points = torch.stack((x_lift, y_lift, z_cam, torch.ones_like(z_cam)), dim=-1)[:, :, :3]
img_rel_points = torch.stack((x_cam, y_cam, z_cam, torch.ones_like(z_cam)), dim=-1)[:, :, :3]
cam_rel_points.shape
```

```python
ray_dir = cam_rel_points
z = ray_dir[:,:,0]	 # z坐标范围-50~50
x = ray_dir[:,:,1]
y = ray_dir[:,:,2]
ddd_points = np.zeros((1,4096,3))
ddd_points[:,:,0] = z
ddd_points[:,:,1] = x
ddd_points[:,:,2] = y
ax = plt.subplot(projection = '3d')  # 创建一个三维的绘图工程
ax.set_title('camera points')  # 设置本图名称
ax.scatter(x, y, z, c = 'r') 
ax.scatter(0, 0, 0, c = 'b') 
```

```python
H = W = 64
K = torch.from_numpy(np.array(
        [[233, 0, 256],
         [0, 233, 256],
         [0, 0, 1]], dtype=np.float32))
i, j = torch.meshgrid(torch.linspace(0, W-1, W), torch.linspace(0, H-1, H))  # pytorch's meshgrid has indexing='ij'
i = i.t()
j = j.t()
dirs = torch.stack([(i-K[0][2])/K[0][0], -(j-K[1][2])/K[1][1], -torch.ones_like(i)], -1)
dirs = dirs.reshape(H*W, 3)
print(dirs.shape)
x = dirs[:,0]	 # z坐标范围-50~50
y = dirs[:,1]
z = dirs[:,2]
ddd_points = np.zeros((4096,3))
ddd_points[:,0] = x
ddd_points[:,1] = y
ddd_points[:,2] = z
ax = plt.subplot(projection = '3d')  # 创建一个三维的绘图工程
ax.set_title('camera points')  # 设置本图名称
ax.scatter(x, y, z, c = 'b') 
ax.scatter(0,0,0,c="r")
```
